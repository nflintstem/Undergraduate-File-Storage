```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eulerplus1(ynow, dt,a,b,t):
  return (ynow + dt*b*(t+dt)) / (1.0 + a*dt)

def trapezoidplusone(ynow,a,b,t,dt):
  return (ynow*(1-a*dt/2.0) + dt*b*(t+dt/2.0)) / (1.0 + a*dt/2.0)
```

```python
t0=0
neq=2
x0=2
y0=1
tmax=30.0
dt=0.01

def gx(t, x, y):
  a=1.2
  b=0.6
  return a*x-b*x*y

def gy(t, x, y):
  c=0.8
  d=0.3
  return -c*y+d*x*y

def g(t, X):
  return np.array([gx(t, X[0], X[1]), gy(t, X[0], X[1])])

nint=int(round((tmax-t0)/dt))
X=np.zeros((neq,nint+1))
t=np.zeros(nint+1)
X[0,0]=x0
X[1,0]=y0

for n in range(nint):
    t[n+1]=t[n]+dt
    X[:,n+1]=X[:,n]+dt*g(t[n], X[:,n])
```

```python
# Plot curve
plt.plot(X[0,:],X[1,:])

# Plot first and last point
plt.plot(x0, y0, '.r', label='_nolegend_')
plt.plot(X[0,-1], X[1,-1], 'xr', label='_nolegend_')

plt.title('$\Delta t=$'+str(dt))
xlabelname='prey'
ylabelname='predator'
plt.xlabel(xlabelname)
plt.ylabel(ylabelname)
plt.show()
```
