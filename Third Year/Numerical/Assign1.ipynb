```python
# Defining Function
def f(x):
    return -0.955968*x**3 - 2.867905*x**2 + 9.559683*x

# Defining derivative of function
def fprime(x):
    return (3*-0.955968)*x**2 - (2*-2.867905)*x + 9.559683

def newtonRaphson(x):
    h = f(x)/fprime(x)
    while abs(h) >= 1e-15:
        h = f(x)/fprime(x)
        x = x - h
    return x

roots = []
i = -50
while i != 50:
  root = newtonRaphson(i)
  if root not in roots:
    roots.append(root)
  i+=10

roots.sort()
print(roots)
```

```python
def DEmat(num):
  mat = np.zeros((n,n))
  for i in range(n):
    mat[i][i] = -2
  for j in range(n-2):
    mat[j+1][j] = 1
    mat[j+1][j+2] = 1
  return mat

def ansMat(n,L,R,ub,lb):
  mat = np.zeros(n)
  mat[0] = -2*L
  mat[-1] = -2*R
  for i in range(1,n-1):
    deltax = xi(n,i,ub,lb)
    if i == 1:
      x = deltax
    else:
      x += (ub-lb)/(n-1)
    mat[i] = g(x) * (deltax**2)
  return mat
```
