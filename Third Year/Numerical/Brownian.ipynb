```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def Brownian():  
  mu, sigma = 0, 1 # mean and standard deviation
  n = 10000 # number of random numbers
  Z_all=np.random.normal(mu, sigma, n) # Generate the random numbers
  #plt.hist(Z_all, 30, density=True) # Show a histogram
  W = np.zeros(n)
  W[0] = 0

  for i in range (1,n):
    W[i] = W[i-1]+Z_all[i-1]
  plt.plot(W)  
```

```python
def OrnUhl(k,dt):
  mu, sigma = 0, 1 # mean and standard deviation
  n = 10000 # number of random numbers
  Z_all=np.random.normal(mu, sigma, n) # Generate the random numbers
  #plt.hist(Z_all, 30, density=True) # Show a histogram
  W = np.zeros(n)
  if dt == 1:
    for t in range(1,n):
      W[t] = (1-k)*W[t-1]+Z_all[t-1]
  else:
    for t in np.arange(1,n,dt):
      W[t] = (1-(k*dt))*W[t-1]+Z_all[t-1]

  plt.plot(W)  
```
